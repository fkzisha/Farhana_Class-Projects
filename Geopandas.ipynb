```python
# Reading a shapefile into geopandas 
# Here I used the California Protected Area Database (CPAD), Dec'22
# In this database, the shapefile of holdings is the individual parcels of protected lands.

shpFileIn = data_location +"CPAD_2022b_Holdings.shp"
```

```python
# Let's explore how different classification methods change the appearance of a map

import mapclassify 

# Specifying classification sheme requires an additional library 
CPAD.plot(column='ACRES', cmap='OrRd', scheme ='equal_interval', k=5, legend=True)
CPAD.plot(column='ACRES', cmap='OrRd', scheme ='fisher_jenks', k=5, legend=True)
CPAD.plot(column='ACRES', cmap='OrRd', scheme ='quantiles', k=5, legend=True)
```
